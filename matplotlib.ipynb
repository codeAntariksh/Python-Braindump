## Bar-Chart

Bi-variate Analysis

Numerical vs Categorical

Use-case= Aggregate Analysis of Groups

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color="yellow") #and all other functions applicable
plt.grid()
plt.show()
```

```python
# Horizontal Bars
# Especially when no.of categories increases
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.barh(colors,children,color="yellow") #and all other functions applicable
plt.grid()
plt.show()
```

## Stacked Bar Chart

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')
plt.legend()
```

```python
# Annotations
batters = pd.read_csv(r"batter.csv")
batters.shape
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,"Point 1",fontdict={'size':12,'color':'red'})
plt.text(2,6,"Point 2",fontdict={'size':12,'color':'brown'})
plt.text(3,7,"Point 3",fontdict={'size':12,'color':'yellow'})
plt.text(4,8,"Point 4",fontdict={'size':12,'color':'green'})
```

```python
# PLOT HORIZONTAL AND VERTICAL LINES
# I want to bid for batsmen whose strike rate is > 130 and avg>30
plt.figure(figsize=(18,10))
plt.axhline(130,color="red")
plt.axvline(30,color="red")
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'],color="red",marker="*")
ax.set_title("Something")
ax.set_xlabel("AVG")
ax.set_ylabel("Strike_Rate")
fig.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(12,7))
ax[0].scatter(batters['avg'],batters['strike_rate'],color="red",)
ax[1].scatter(batters['avg'],batters['runs'],color="violet")

ax[0].set_title('Avg Vs Strike-Rate')
ax[0].set_ylabel("Strike Rate")
ax[0].set_xlabel("Avg")

ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0][0].scatter(batters['avg'],batters['strike_rate'],color="red")
ax[0][1].scatter(batters['avg'],batters['runs'],color="violet")

ax[1][0].hist(batters['avg'],color="red")
ax[1][1].hist(batters['runs'],color="violet")

# Again Individually can be optimised
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color="red")
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'],color="violet")
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'],color="red")
ax4 = fig.add_subplot(2,2,4)
ax4.hist(batters['runs'],color="violet")

# Again Individually can be optimised
```

```python
# 3D Surface Plots
plt.style.use("bmh")
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection="3d")
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
s = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(s)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection="3d")
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
xx,yy = np.meshgrid(x,y)
z = np.sin(xx) + np.cos(yy)
s = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(s)
```

```python
# Contour PLot
fig = plt.figure(figsize=(15,15))
ax = plt.subplot()
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
p = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot()
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
xx,yy = np.meshgrid(x,y)
z = np.sin(xx) + np.cos(yy)
s = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(s)
```
