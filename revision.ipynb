```python
# Functions
def assessment(phys,chem,maths):
    total = phys+chem+maths
    avg = (total)/3
    if(avg>=90):
        grade = "A"
    elif(avg>=80 and avg<90):
        grade = "B"
    elif(avg>=50 and avg<80):
        grade = "P"
    else:
        grade = "F" 
    return total,avg,grade
```

```python
total,avg,grade = assessment(80,95,100)
print(total,avg,grade)
```
