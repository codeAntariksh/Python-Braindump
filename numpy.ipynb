### Broadcasting

The term Broadcasting describes how NumPy treats arrays with different shapes during Arithmetic Operations

The smaller array is Broadcasted across a larger array so that they have compatible shapes

Kind off like Taking a Convolution

```python
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a+b)
```

### Broadcasting Rules

##### 1. Make the 2 Arrays have the same no. of dimensions
If the number of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.

##### 2. Make each dimension of the two arrays same in size
If the sizes of the each dimensions of the 2 Arrays do not match. dimensions with size 1 are stretched to the size of the other array.
If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted.

```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)
print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)
print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)
print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)
print(a+b)
```

Numpy works in process of Vectorization and Boradcasting also is done on the basis of that.
Also this is the reason NumPy works very fast

Usage of Any Formula with Numpy Arrays

```python
# Sigmoid
def sigmoid(array):
    return 1/(1+np.exp(-(array)))

a = np.arange(10)
sigmoid(a)
```

```python
# ReLU
def ReLU(array):
    return np.where(array>=0,array,0)

a = np.arange(15)
ReLU(a)
```

```python
# Mean Squared Error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
def mse(actual,predicted):
    return np.mean((actual-predicted)**2)

mse(actual,predicted)
```

```python
actual = np.random.random((1,10))
predicted = np.random.random((1,10))
# Binary Cross Entropy
def bce(actual,predicted):
    return (-(np.mean((actual*np.log(predicted)) + (1-actual)*np.log(1-predicted))))

print(bce(actual,predicted))
```
