```python
# General Logic on if-else,Looping,OOPS (Basic) & Functions
class Student:

  def calculate_avg(self):
    avg = 0
    for i in range(len(self.marks)):
      avg+=self.marks[i]
    avg/=len(self.marks)
    return avg

  def assign_grade(self):
    if(self.avg>=90):
      return "A"
    elif (self.avg>=75):
      return "B"
    elif(self.avg>=60):
      return "C"
    else:
      return "D"

  def has_failed(self):
    for i in range(len(self.marks)):
      if(self.marks[i]<40):
        return True
    else:
      return False

  def display_details(self):
    print("Name:{}".format(self.name))
    print("Marks:{}".format(self.marks))
    print("Avg:{}".format(self.avg))
    print("Has Failed:{}".format(self.has_failed()))
    print("Grade:{}".format(self.grade))

  def __init__(self,roll,marks,name):
    self.name = name
    self.roll = roll
    self.marks = marks
    self.avg = self.calculate_avg()
    self.grade = self.assign_grade()
```

```python
# Usage of Modules, importing coupled with logic and OOPS
%%writefile geometry.py
import math

class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def diagonal(self):
        return math.sqrt(self.length**2 + self.width**2)

    def is_square(self):
        if self.length == self.width:
            return True
        else:
            return False

```

```python
## Testing on Functions(multiple Returns), Args and kwargs, Built in Attributes
def analyze_numbers(*args, operation="all", **kwargs):
  absolute = kwargs.get("absolute", False)
  round_to = kwargs.get("round_to", None)

  def summation(*args,**kwargs):
    summer = 0
    for i in range(len(args)):
      summer = summer + args[i]
    if(kwargs["absolute"]==True):
      summer = abs(summer)
    if(kwargs["round_to_int"]):
      summer = round(summer,kwargs["round_to_int"])
    return summer

  def avg(*args):
    sum = summation(*args)
    return sum/len(args)
  def minmax(*args):
    mini = 0
    maxi = 0
    if(kwargs["absolute"]==True):
      mini = abs(min(*args))
      maxi = abs(max(*args))
    if(kwargs["round_to_int"]):
      mini = round(min(*args),kwargs["round_to_int"])
      maxi = round(max(*args),kwargs["round_to_int"])
    return mini,maxi
  def all(*args):
    return summation(args),avg(args),minmax(args)

  if(operation=="all"):
    return all(*args)
  elif(operation=="avg"):
    return avg(*args)
  elif(operation=="minmax"):
    return minmax(*args)
  elif(operation=="summation"):
    return summation(*args)
  else:
    return None

```

```python
def analyze_numbers(*args, operation="all", **kwargs):
    absolute = kwargs.get("absolute", False)
    round_to = kwargs.get("round_to", None)

    numbers = []

    for value in args:
        if isinstance(value, (int, float)):
            if absolute:
                value = abs(value)
            numbers.append(value)

    if len(numbers) == 0:
        return None

    total = 0
    minimum = numbers[0]
    maximum = numbers[0]

    for num in numbers:
        total += num
        if num < minimum:
            minimum = num
        if num > maximum:
            maximum = num

    average = total / len(numbers)

    if round_to is not None:
        total = round(total, round_to)
        average = round(average, round_to)
        minimum = round(minimum, round_to)
        maximum = round(maximum, round_to)

    if operation == "sum":
        return total
    elif operation == "avg":
        return average
    elif operation == "minmax":
        return (minimum, maximum)
    elif operation == "all":
        return (total, average, minimum, maximum)
    else:
        return None
```
