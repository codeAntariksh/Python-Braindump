# Seaborn

Another Data Visualisation Library (like matplotlib,plotly,etc)

Why Seaborn?


1. Provides a layer of abstraction hence simpler to use

2. Better Aesthetics

3. More Graphs included

Seaborn Roadmap

Types of Functions
1. Figure Level
2. Axis Level

Main Classification

1. Relational PLot (Scatterplot & lineplot)
2. Distribution Plot (Histgram,Rug Plot and KDE Plot)
3. Categorical Plot (Bar Plots,Count Plot,Box Plot,Violin Plot,Swamplot)
4. Matrix Plot (Heatmaps, Cluster Map)
5. Regression plots ()
6. Multiplots (Joint Plot, Pair Plot)

A Graph/Plotting has a figure level and an underlying axes levels which are used to paint the whole picture

Hence 2 types of functions-> Figure and Axis Level

### Box Plot

In Box Plots,
The Box represents the %ile distribution from 25th to 75th %ile (lower and upper vertex signified) and also a calculated minimum (at lower end of line) & a calculated maximum (at higher end of line) with external or otherwise called Outliers.

In Violin Plot,
Box Plot combined with a width sort of interface signifying KDE Plot involving Probabilistic Definitions is generally drawn.

```python
sns.boxplot(data=tips,x="sex",y="total_bill")
```

```python
sns.catplot(kind="box",data=tips,x="day",y="total_bill",hue="sex")
```

```python
# For a single column representation
sns.boxplot(data=tips,x="total_bill")
```

Violinplot = BoxPLot + KDE Plot

```python
sns.violinplot(data=tips,x="day",y="total_bill",color="violet")
```

```python
sns.catplot(kind="violin",data=tips,x="day",y="total_bill",hue="sex")
```

```python
sns.catplot(kind="violin",data=tips,x="day",y="total_bill",hue="sex",split=True)
```
