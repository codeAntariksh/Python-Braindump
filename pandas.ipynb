```python
points_df = pd.DataFrame(
    {
        '1st Point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd Point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def eucld(row):
    return distance(row['1st Point'],row['2nd Point'])

def distance(val1,val2):
    return np.sqrt((val1[0]-val2[0])**2 + (val1[1]-val2[1])**2)

points_df['Euclidean Distance'] = points_df.apply(eucld,axis=1)
```
